<h5>Análise gráfica dos resultados</h5>

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
import plotly.io as pio
pio.renderers.default = "browser"
```

```python
data_16 = pd.read_csv("dados/acc_score_16ch.csv", sep=';')
data_8 = pd.read_csv("dados/acc_score_8ch.csv", sep=';')
data_4 = pd.read_csv("dados/acc_score_4ch.csv", sep=';')
data_2 = pd.read_csv("dados/acc_score_2ch.csv", sep=';')
```

```python
data_2
```

```python
df_acc = pd.DataFrame({'n_eletrodos':[2,4,8,16], 'Acurácia':[data_2.acc_media.iloc[-1],data_4.acc_media.iloc[-1],data_8.acc_media.iloc[-1],data_16.acc_media.iloc[-1]],
'std':[data_2.acc_media.iloc[0:19].std(),data_4.acc_media.iloc[0:19].std(),data_8.acc_media.iloc[0:19].std(),data_16.acc_media.iloc[0:19].std()]})
df_acc
```

```python
plt.figure(figsize=(15,9))
plt.plot(df_acc['n_eletrodos'], df_acc['Acurácia'],'k',linewidth=1.3)
plt.xlabel('Número de eletrodos')
plt.ylabel('Acurácia média (%)')
plt.title('Relação entre a acurácia média e o número de eletrodos')
plt.grid()
plt.show()
```
